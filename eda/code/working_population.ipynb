# 직장인구_자치구 전처리

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager as fm
from matplotlib.ticker import StrMethodFormatter
import sys

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
df = pd.read_csv('../../data/직장인구_자치구.csv')
```

```python
df = df.drop(['자치구_코드'], axis=1)
```

```python
df.head()
```

```python
ignore_cols = [c for c in df.columns if c != "기준_년분기_코드"]

total_rows = len(df)
unique_rows_ignoring_code = len(df[ignore_cols].drop_duplicates())

print(f"전체 행수: {total_rows:,}")
print(f"'기준_년분기_코드'를 제외한 고유 행수: {unique_rows_ignoring_code:,}")
```

```python
grp = (df
       .groupby(ignore_cols, as_index=False)["기준_년분기_코드"]
       .nunique()
       .rename(columns={"기준_년분기_코드": "분기코드_개수"}))

only_code_diff = grp[grp["분기코드_개수"] > 1]

print(f"'기준_년분기_코드'만 달라서 중복인 묶음 개수: {len(only_code_diff):,}")

if len(only_code_diff) > 0:
    sample_key = only_code_diff.iloc[0][ignore_cols]
    ex = df.merge(only_code_diff.head(1), on=ignore_cols, how="inner")
    display(ex.head(10))
```

```python
df.sort_values('기준_년분기_코드', inplace=True)
df.to_csv('./../data/working_population_data.csv')
```

### 시각화

```python
male_total_col, female_total_col = "남성_직장인구_수", "여성_직장인구_수"
male_age_cols   = [f"남성연령대_{x}_직장인구_수" for x in ["10","20","30","40","50","60_이상"]]
female_age_cols = [f"여성연령대_{x}_직장인구_수" for x in ["10","20","30","40","50","60_이상"]]

def get_total(total_col, fallback_cols):
    if total_col in df.columns:
        return df[total_col].sum()
    cols = [c for c in fallback_cols if c in df.columns]
    return df[cols].sum().sum() if cols else 0

male_total   = get_total(male_total_col, male_age_cols)
female_total = get_total(female_total_col, female_age_cols)

fig, ax = plt.subplots(figsize=(6,5))
ax.bar(["남성_총계","여성_총계"], [male_total, female_total], color=["tab:blue", "#ff69b4"])
ax.set_title("남녀 직장인구수 총계 비교")
ax.set_ylabel("총합")

ax.yaxis.set_major_formatter(StrMethodFormatter("{x:,.0f}"))

plt.tight_layout()
plt.show()
```

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter

try:
    df
except NameError:
    df = pd.read_csv("직장인구_자치구.csv", encoding="utf-8-sig")

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

labels = ["10대","20대","30대","40대","50대","60대 이상"]

male_cols = [
    "남성연령대_10_직장_인구_수",
    "남성연령대_20_직장_인구_수",
    "남성연령대_30_직장_인구_수",
    "남성연령대_40_직장_인구_수",
    "남성연령대_50_직장_인구_수",
    "남성연령대_60_이상_직장_인구_수",
]
female_cols = [
    "여성연령대_10_직장_인구_수",
    "여성연령대_20_직장_인구_수",
    "여성연령대_30_직장_인구_수",
    "여성연령대_40_직장_인구_수",
    "여성연령대_50_직장_인구_수",
    "여성연령대_60_이상_직장_인구_수",
]

missing = [c for c in male_cols + female_cols if c not in df.columns] # 누락검사
if missing:
    raise KeyError(f"다음 컬럼이 없습니다:\n{missing}")

M = [pd.to_numeric(df[c], errors="coerce").sum() for c in male_cols]
F = [pd.to_numeric(df[c], errors="coerce").sum() for c in female_cols]

x = np.arange(len(labels)); w = 0.6
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(x, M, width=w, label="남성", color="tab:blue")
ax.bar(x, F, width=w, bottom=M, label="여성", color="#ff69b4")

ax.set_title("연령대별 남녀 직장인구수 (스택형)")
ax.set_xlabel("연령대"); ax.set_ylabel("총합")
ax.set_xticks(x); ax.set_xticklabels(labels)

ax.yaxis.set_major_formatter(StrMethodFormatter("{x:,.0f}"))
```
