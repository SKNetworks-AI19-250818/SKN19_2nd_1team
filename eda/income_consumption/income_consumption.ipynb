```python
df = df.drop(['자치구_코드', '소득_구간_코드'], axis=1)
```

```python
df.head()
```

```python
df.sort_values('기준_년분기_코드', inplace=True)
df.to_csv('income_consumption_data.csv')
```

```python
df['지출_총_금액'].plot(kind='box')
plt.show()
```

```python
df_filtered = df[df['기준_년분기_코드'] == 20252]

plt.bar(df_filtered['자치구_코드_명'], df_filtered['지출_총_금액'])
plt.xticks(rotation=55)
plt.show()
```

```python
df['월_평균_소득_금액'].plot(kind='box')
plt.show()
```

```python
df_filtered = df[df['기준_년분기_코드'] == 20252]

plt.bar(df_filtered['자치구_코드_명'], df_filtered['월_평균_소득_금액'])
plt.xticks(rotation=55)
plt.show()
```
